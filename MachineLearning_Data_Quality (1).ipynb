# Function that creates the training arrays (increases the number of training sets)

```python
def create_w2p_training_set(pretrain_folder,number_training_sets=1000,expansion_factor=1):
    """Given a directory containing multiple measurements this creates a list of np.arrays to 
    train a classifier"""
    #sets the path to the files that will train the classifier 
    pretrain_files=os.listdir(pretrain_folder)
    pretrain_files=map(lambda x:os.path.join(pretrain_folder,x),pretrain_files)
    # now open the files convert them to AsciiData Tables and find the mean and standard deviation
    first_file=W2P(pretrain_files[0])
    joined_file=first_file.copy()
    for file_name in pretrain_files[1:]:
        joined_file=joined_file+W2P(file_name)
    #calculates the mean and standard deviation on the joined files 
    mean_file=frequency_model_collapse_multiple_measurements(joined_file)
    std_file=frequency_model_collapse_multiple_measurements(joined_file,method="std")
    mean_A1_D1=[complex(re_A1_D1,mean_file["imA1_D1"][index])
            for index,re_A1_D1 in enumerate(mean_file["reA1_D1"])]
    #joins the magnitude and phase of A1_D1 in order to use to train the classifier  
    std_A1_D1=np.abs(np.array([complex(re_A1_D1,std_file["imA1_D1"][index]) for index,re_A1_D1 
                           in enumerate(std_file["reA1_D1"])]))
    #increases the number of training sets
    training_arrays=np.array([np.random.normal(loc=a1_d1,scale=expansion_factor*std_A1_D1[index]+.0001,
                                               size=number_training_sets) for index,
                              a1_d1 in enumerate(np.abs(np.array(mean_A1_D1)))])

    #reorganizes the array 
    training_arrays=[[training_arrays[i][j] 
                      for i in range(len(mean_A1_D1))] for j in range(number_training_sets)]
    return training_arrays
```

# Creates the training array and target array for the classifier. 

```python
number_training_sets=1000
expansion_factor=1
sorted_device_keys=sorted(quality_type_dictionary.keys())
training_array=[]
for key in sorted_device_keys:
    training_array=training_array+create_w2p_training_set(pretrain_folder,number_training_sets,expansion_factor =2)
    
target_array=[]
for key_index,key in enumerate(sorted_device_keys):
    target_array=target_array+[key for i in range(number_training_sets)]
```
